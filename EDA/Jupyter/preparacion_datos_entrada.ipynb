### En este Jupyter Notebook preparé los archivos con los datos que recogen la información de los vehículos retirados de la vía pública de Madrid, desde 2016 hasta 2023.

Los archivos, en formato CSV y uno para cada año, tienen las siguientes columnas:

- **fechaEntrada**: Indica la fecha y hora en que se retiró el vehículo de la vía publica.
- **tipoVia**: Indica el tipo de vía desde la cual se retiró el vehículo.
- **Via**: Nombre de la vía.
- **numero**: Número de la vía.
- **cp**: Código postal de la vía.
